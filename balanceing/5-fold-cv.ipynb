```python
M_df = df[df["diagnosis"] == "M"]
B_df = df[df["diagnosis"] == "B"]
```

```python
# Sample 57 instances without replacement
M_test = M_df.sample(n=72, replace=False, random_state=42)
B_test = B_df.sample(n=72, replace=False, random_state=42)

test_concat = pd.concat([M_test, B_test])

# Drop the sampled instances from the original dataframe
M_df = M_df.drop(M_test.index)
B_df = B_df.drop(B_test.index)

# Verify that the sampled instances were removed from the original dataframe
print(M_df.shape)
print(B_df.shape) 
```

```python
test_concat.to_csv("balance_data/test.csv", index=False)
```

```python
# Sample 57 instances without replacement
M_fold = M_df.sample(n=57, replace=False, random_state=42)
B_fold = B_df.sample(n=57, replace=False, random_state=42)

fold_concat = pd.concat([M_fold, B_fold])

# Drop the sampled instances from the original dataframe
M_df = M_df.drop(M_fold.index)
B_df = B_df.drop(B_fold.index)

# Verify that the sampled instances were removed from the original dataframe
print(M_df.shape)
print(B_df.shape)

test_concat.to_csv("balance_data/fold5.csv", index=False)
```
