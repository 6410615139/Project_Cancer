```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('raw_data/data.csv')
df.head()
```

```python
list(df.columns)
```

```python
df.isnull().sum()
```

```python
df = df.drop(labels='Unnamed: 32', axis=1)
df.head()
```

```python
df.info()
```

```python
data_mean = df[['id','diagnosis','radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','symmetry_mean','fractal_dimension_mean']]
data_mean.info()
```

```python
# select the columns to calculate correlation with 'diagnosis'
cols = df.columns[2:]

# calculate the correlation between 'diagnosis' column and other columns
corr_matrix = df[cols].corrwith(df['diagnosis'].apply(lambda x: 1 if x == 'M' else 0))

# sort the correlation values in descending order
corr_matrix_sorted = corr_matrix.sort_values(ascending=True)

# create a bar plot of the sorted correlation values
plt.bar(corr_matrix_sorted.index, corr_matrix_sorted)

# set the axis labels and title
plt.xlabel('Features')
plt.ylabel('Correlation with Diagnosis')
plt.title('Correlation between Diagnosis and Features')

# rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# show the plot
plt.show()
```

```python
# select the columns to include in the pairplot
cols = ['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean']

# Create a dictionary to map the diagnosis values to colors
color_map = {'M': 'r', 'B': 'g'}

# create a pairplot of the selected columns
sns.pairplot(df[cols], hue='diagnosis', palette=color_map)

# show the plot
plt.show()
```
