```python
import pandas as pd
```

```python
# Read the two CSV files into separate DataFrames
f1 = pd.read_csv("../SVM_progress3/DATA/f1.csv")
f2 = pd.read_csv("../SVM_progress3/DATA/f2.csv")
f3 = pd.read_csv("../SVM_progress3/DATA/f3.csv")
f4 = pd.read_csv("../SVM_progress3/DATA/f4.csv")
f5 = pd.read_csv("../SVM_progress3/DATA/f5.csv")
```

```python
# set up path 
data_path = "../SVM_progress3/DATA/"

# create train set
train_set = pd.concat([f1,f2,f3,f4,f5])
train_set.to_csv(data_path+"train.csv", index=False)
```

```python
def create_split(split_num):
    split = "split{}".format(split_num)
    val = pd.read_csv("../SVM_progress3/DATA/f{}.csv".format(split_num))
    num_5 = list(range(1,6))
    num_5.remove(split_num)
    first = pd.read_csv("../SVM_progress3/DATA/f{}.csv".format(num_5[0]))
    second = pd.read_csv("../SVM_progress3/DATA/f{}.csv".format(num_5[1]))
    third = pd.read_csv("../SVM_progress3/DATA/f{}.csv".format(num_5[2]))
    fourth = pd.read_csv("../SVM_progress3/DATA/f{}.csv".format(num_5[3]))
    train = pd.concat([first, second, third, fourth])
    return train, val
```

```python
train_1, val_1 = create_split(1)
```
