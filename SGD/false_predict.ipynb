```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

```python
def plot_scatter(split, false_indices):
    path = "../kitt/DATA_progress3/{}/val/val.csv".format(split)
    val_data = pd.read_csv(path)
    
    x_col = 'area_mean'
    y_col = 'concavity_mean'

    # Get the false predictions data from the original data using the indices of false predictions
    false_data = val_data.iloc[false_indices]

    # Get the true predictions data from the original data using the indices of true predictions
    malignant_data = val_data[57:]
    benign_data = val_data[:56]

    # Create the scatter plot with false predictions data as red dots and true predictions data as blue dots
    plt.scatter(benign_data[x_col], benign_data[y_col], color='green', label='B')
    plt.scatter(malignant_data[x_col], malignant_data[y_col], color='blue', label='M')
    plt.scatter(false_data[x_col], false_data[y_col], color='red', label='False Predictions')

    # Add legend and labels
    plt.legend()
    plt.xlabel(x_col)
    plt.ylabel(y_col)

    # Show the plot
    plt.show()
```

```python
def swarm(split, false_indices):
    # train_path = "../kitt/DATA_progress3/split1/train/train.csv"
    # train_data = pd.read_csv(path)
    val_path = "../kitt/DATA_progress3/{}/val/val.csv".format(split)
    val_data = pd.read_csv(path)
    
    x_col = 'area_mean'
    x = val_data[x_col]
    y = val_data.diagnosis 
    # f = val_data.iloc[[20]]
    
    sns.set(style="whitegrid", palette="muted")
    data_dia = y
    data = x
    data_n_2 = (data - data.mean()) / (data.std())
    data = pd.concat([y,data_n_2.iloc[:]],axis=1)
    data = pd.melt(data,id_vars="diagnosis",
                        var_name="features",
                        value_name='value')
    # f = pd.melt(f,id_vars="diagnosis",
    #                     var_name="features",
    #                     value_name='f')
    plt.figure(figsize=(5,5))
    tic = time.time()
    sns.swarmplot(x="features", y="value", hue="diagnosis", data=data)
    # sns.swarmplot(x="features", y="f", hue="diagnosis", data=f)

    plt.xticks(rotation=90)
```
